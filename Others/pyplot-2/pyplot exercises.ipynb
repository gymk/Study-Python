```python
time = [0, 1, 2, 3, 4, 5, 6]
CO2_ppm = [250, 265, 272, 260, 300, 320, 389]

plt.plot(time, CO2_ppm, 'b--')
plt.xlabel('time')
plt.ylabel('CO2 ppm')
plt.title('C02 Level chart')
plt.show()
```

```python
Temp = [14.1, 15.5, 16.3, 18.1, 17.3, 19.1, 20.2]
fig = plt.figure()
plt.plot(time, CO2_ppm, 'b--', time, Temp, 'r*-')
plt.xlabel('time')
plt.ylabel('CO2 ppm')
plt.show()
fig.savefig('c02_temp.pdf')
```
