```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2 * np.pi, 600)
y = np.sin(x ** 2)
```

```python
plt.close('all')
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.close('all')
```

```python
f, ax = plt.subplots()
print(type(f),type(ax))
ax.plot(x, y)
ax.set_title('Simple Plot example')
f.show()
plt.show()
```
